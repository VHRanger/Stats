```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import numpy as np
import statsmodels.api as sm
%matplotlib inline
```

```python
data = pd.read_csv('c:/users/matt/desktop/poker paper/sessions_dummies.csv')
```

```python
data['Session Start'] = pd.to_datetime(data['Session Start'])
data = data.sort('Session Start')
data = data.drop('Unnamed: 0', axis=1)
data = data.reset_index()
```

```python
X = data.loc[:, 'Treat':'Time']
X = sm.add_constant(X)
# X = X.drop('After', axis=1)
minutes = data[['Minutes']]
logminutes = data['logminutes']
```

```python
est1 = sm.OLS(minutes, X).fit()
est1.summary()
```

```python
est2 = sm.OLS(logminutes, X).fit()
est2.summary()
```
