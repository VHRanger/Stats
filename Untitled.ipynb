```python
import pandas as pd
from io import StringIO

f = "20201206_17h33m25_Waveform.txt"

with open(f, 'r',encoding='latin1') as ff:
    lines = ff.readlines()

# Prendre les noms de la premiere ligne
noms = lines[0].split()

# Joindre le reste des lignes en dataset
dataset = lines[1:]
dataset = '\n'.join(dataset)
df = pd.read_csv(StringIO(dataset), encoding='latin1', delim_whitespace=True)
# enleve les colonnes vides
df = df.dropna(axis=1, how='all')
```

```python
noms_quadruple = []
for n in noms:
    noms_quadruple.append(n)
    noms_quadruple.append(n)
    noms_quadruple.append(n)
    noms_quadruple.append(n)

nouvelles_colonnes = []

for i in range(len(df.columns)):
    nouvelles_colonnes.append(f"{noms_quadruple[i]}_{df.columns[i]}")

df.columns = nouvelles_colonnes
```

```python
import seaborn as sns
sns.set()

sns.pairplot(df[["Residus_Time", "Residus_(s)", "Residus_Long", "Residus_(mm/s)"]])
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(df['Residus_(s)'], model='additive', period=1)
result.plot()
```
